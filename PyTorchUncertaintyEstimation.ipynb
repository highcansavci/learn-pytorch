```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
```

```python
def generate_batch(batch_size=32):
  # (-5, 5)
  x = np.random.random(batch_size) * 10 - 5

  # sd is a function of x
  sd = 0.05 + 0.1 * (x + 5)

  # target = mean + noise * sd
  y = np.cos(x) - 0.3 * x + np.random.randn(batch_size) * sd

  return x, y
```

```python
# visualize the data
x, y = generate_batch(1024)
plt.scatter(x, y, alpha=0.5);
```

```python
class Model(nn.Module):
  def __init__(self):
    super().__init__()

    self.ann_mean = nn.Sequential(
        nn.Linear(1, 10),
        nn.Tanh(),
        nn.Linear(10, 1)
    )

    self.ann_sd = nn.Sequential(
        nn.Linear(1, 10),
        nn.Tanh(),
        nn.Linear(10, 1)
    )

  def forward(self, inputs):
    # returns (mean, log-variance)
    return self.ann_mean(inputs), self.ann_sd(inputs)
```

```python
model = Model()
```

```python
def criterion(outputs, targets):
  mu = outputs[0]
  v = torch.exp(outputs[1])

  # coefficient term
  c = torch.log(torch.sqrt(2 * np.pi * v))

  # exponent term
  f = 0.5 / v * (targets - mu) ** 2

  # mean log-likelihood
  nll = torch.mean(c + f)

  return nll
```

```python
optimizer = torch.optim.Adam(model.parameters())
```

```python
n_epochs = 5000
batch_size = 128
losses = np.zeros(n_epochs)
for i in range(n_epochs):
  x, y = generate_batch(batch_size)

  # convert to torch tensor
  inputs = torch.from_numpy(x).float()
  targets = torch.from_numpy(y).float()

  # reshape the data
  inputs = inputs.view(-1, 1)
  targets = targets.view(-1, 1)

  # zero grad
  optimizer.zero_grad()

  # forward pass
  outputs = model(inputs)
  loss = criterion(outputs, targets)

  # store loss
  losses[i] = loss.item()

  # print loss
  if i % 1000 == 0:
    print(i, losses[i])
  
  # optimize
  loss.backward()
  optimizer.step()
```

```python
plt.plot(losses);
```

```python
# plot the model predictions
x, y = generate_batch(1024)
plt.scatter(x, y, alpha=0.5)

# convert to torch tensor
inputs = torch.from_numpy(x).float()
targets = torch.from_numpy(y).float()

# reshape data
inputs, targets = inputs.view(-1, 1), targets.view(-1, 1)

with torch.no_grad():
  outputs = model(inputs)
  yhat = outputs[0].numpy().flatten()
  sd = np.exp(outputs[1].numpy().flatten()) # since encoded variance by default

idx = np.argsort(x)
plt.plot(x[idx], yhat[idx], linewidth=3, color="red")
plt.fill_between(x[idx], yhat[idx] - sd[idx], yhat[idx] + sd[idx], color="red", alpha=0.5);
```
